```python
#Cyclinder
vehical_data.cylinders.value_counts()
vehical_data.cylinders.value_counts().sum()
#barplot
sns.countplot(vehical_data.cylinders)
plt.ylabel('No of cars')
plt.title('cylinder')
plt.show()
```

```python
vehical_data['acceleration'].plot.box(color='red',vert=False, patch_artist=True)
plt.xlabel('Acceleration')
plt.title('Boxplot-acceleration')
plt.show()
           
```

```python
sns.heatmap(vehical_data.corr(),annot=True)
vehical_data.corr()['acceleration']
```

```python
plt.hist(vehical_data.displacement)
plt.xlabel('Displacement')
plt.ylabel('No of Cars')
plt.title('Histogram- Displacement')
plt.show()
```

```python
sns.scatterplot(x='displacement',y='cylinders',data=vehical_data,hue='cylinders')
plt.title('DISPLACEMENT WITH RESPECT TO CYLINDERS')
plt.show()
```

```python
plt.hist(vehical_data.acceleration)
plt.xlabel('Acceleration')
plt.xticks (rotation=90)
plt.ylabel('No of Cars')
plt.title('Histogram- Acceleration')
plt.show()
```
